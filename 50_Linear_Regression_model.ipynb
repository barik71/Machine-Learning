```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement1.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y ,c = 'r')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train , y_test = train_test_split(x,y,train_size = 0.2,random_state=2)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(x,y)
plt.plot(x_test,lr.predict(x_test),c = 'r')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
```

```python
m = lr.coef_
```

```python
c = lr.intercept_
```

```python
m * 7.15 + c
```
